# Orbitaler i Generel Relativitetsteori
## Introduktion

I denne opgave vil vi regne på, og illustrere, orbitaler af små masser om større masser i lyset af Klassisk Mekanik og Generel Relativitetsteori. Tyngdekraften fra en masse $M$ på en anden masse $m$ er beskrevet på to lidt forskellige måder i de to teorier, i Klassisk Mekanik har vi formlen:

$$ \overrightarrow{F}_{CL} = - \frac{G M m}{r^3} \, \overrightarrow{r}, $$

hvor
- $G \approx 6.67430 \cdot 10^{-11} \mathrm{\frac{m^3}{kg \, s^2}}$ er Newtons tyngdekonstant.
- $\overrightarrow{r}$ er retningsvektoren der pejer fra centeret af $M$ til centeret af $m$.
- $r$ er længden af $\overrightarrow{r}$.

Generel Relativitetsteori tilføjer en lille *rettelse* i lyset af postulatet om at lysets hastighed $c$ er ens for alle observatører. Her Har vi formlen

$$ \overrightarrow{F}_{GR} = - \frac{G M m}{r^3} \left(1 + \frac{3 l^2}{c^2 r^2} \right) \, \overrightarrow{r},$$

hvor
- $G$, $M$, $m$, $r$ og $\overrightarrow{r}$ er det samme som ovenfor.
- $l$ er det angulære moment per masse, givet ved $\left\lVert \overrightarrow{v} \times \overrightarrow{r} \right\rVert$.
- $\overrightarrow{v}$ er hastighedsvektoren hørende til $m$.
- $ c = 299\,792\,458\,\mathrm{\frac{m}{s}}$ er lysets hastighed.

## Enheder og variable
Vi vil arbejde i *naturlige enheder*. Det vil sige at vi vælger at sætte

$$G = c = \hbar = k_B = 1$$

Hvor $G$ og $c$ er beskrevet ovenfor, $\hbar$ er den reducerede Planck konstant $\hbar = \frac{h}{2 \pi}$ og $k_B$ er Boltzmanns konstant. Dette enhedssystem kaldes også for Planck enheder, fordi de først blev foreslået af den tyske fysiker Max Planck i 1899. De to sidste ($\hbar$ og $k_B$) vil vi ikke fokusere mere på i denne opgave, og derfor kan vi nøjes med at betragte enhedssystemet hvor 

$$G = c = 1,$$

også kaldet *geometriserede* enheder.

Yderligere vil vi angive afstande i enheder af Schwarzschild-radiusen, som er givet ved

$$R_S = \frac{2 M G}{c^2} \qquad \text{eller} \qquad R_S = 2 M \quad \text{i naturlige enheder}.$$

Denne radius, som først blev udledet af den tyske astronom Karl Schwarzschild i hans løsning af Generel Relativitetsteori i 1916, betegner den største afstand fra en masse hvor ingen begivenhed, information eller objekt (ikke engang lys) kan undslippe massens tyngdekraft (derfor også kaldet *begivenhedhorisonten*). Når vi skal kigge på orbitaler vil det derfor være irrelevant at fokusere på noget inden for denne radius.

For at kunne angive afstande kun i enheder af Schwarzschild-radiusen er vi nødt til at definere følgende dimensionsløse variable, for at 'normalisere' enhederne kan man sige.

$$\rho = \frac{r}{R_S}, \qquad T = \frac{\tau}{R_S} \quad \text{og} \quad \lambda = \frac{l}{R_S}$$

Hvor $\tau$ er *egentiden*, som har enheden masse.  De skal *repræsentere* hhv. position, tid og angulært moment i vores udregninger.

### Spørgsmål 1

Man kan selvfølgelig spørge sig selv hvorfor vi nu lige pludselig vælger at regne problemet på denne måde, men det har faktisk et par gennemgribende fordele at formulere problemer på denne måde:

- **Simpelhed**

Den 'simpleste' fordel ved at normalisere de naturlige konstanter $G$, $c$ osv. på denne måde er, at mange af udtrykkende som vi bruger til at beskrive fysiske systemer, bliver væsentligt simplere. F.eks. formlen for Schwarzschild-radiusen som beskrevet ovenfor.


- **Tættere knyttet til fundamentale fysiske love**

En anden fordel er at enhederne for masse, afstand og tid nu ikke er defineret af f.eks. hhv. tyngdekraften på et lod, længden af et skridt, eller en anden varierende måling, men ud fra naturens love. Dette gør at der fjernes et abstraktionsniveau når man prøver at forstå den fysiske betydning af et udtryk man regner på. Vi kan f.eks. beregne Solens Schwarzschild-radius i SI-enheder ud fra formlen


$$ R_0 = \frac{2 M G}{c^2} $$ 


Hvor $M = 1.98847 \cdot 10^{30}$ kg er solens masse.

```python
from scipy.constants import G, c

M = 1.98847e+30  # Solens masse i kg
G = G            # Newtons tyngdekonstant på 6.67430e-11 m^3/(kg s)
c = c            # lysets hastighed på 299 792 458 m/s

R0 = 2*M*G/c**2  # Beregning af Solens Schwarzhild-radius i SI-enheder

print(f'Solens Schwarzhild-radius i SI-enheder er ca. {int(R0)} m')
```

Nu kan vi i stedet prøve at udregne Solens Schwarzschild-radius med de naturlige enheder. Vi indsætter $G = c = 1$ i vores formel fra før og får

$$ R_0 = \frac{2 M G}{c^2} = 2 M $$

```python
M = 1.98847e+30  # Solens masse i kg

R0 = 2*M         # Beregning af Solens Schwarzhild-radius i naturlige enheder

print(f'Solens Schwarzhild-radius i naturlige enheder er ca. {R0} kg')
```

Det kan tydeligt ses at Schwarzschild-radiusen er noget der associeret med enhver masse i universet, og at den er ligefrem proportional med massen, med en proportionalitetsfaktor på 2. De naturlige enheder skifter altså perspektivet hvormed vi ser på fysiske problemer (f.eks. hvad er vores fundne værdi ift. lysets hastighed, og ikke bare ift. en arbitrær enhed for fart som m/s).

En finurlig bivirkning ved de naturlige enheder er dog, at enheder på andre kendte størrelser bliver ændret. f.eks. kan vi se at Schwarzschild-radiusen nu har enheden masse. Dette er grunden til at vi jhar indført de tre dimensionsløse variable $\rho$, $T$ og $\lambda$.

Disse variable vil gøre os i stand til at regne på problemet med de nye naturlige variable uden at få problemer med enhederne.

## Beskrivelse af systemet
Med disse nye enheder og variable kan vores system nu beskrives ved ligningen for energitætheden af massen $m$

$$ \mathcal{E} = \frac{1}{2} \left(\frac{\mathrm{d} \rho}{\mathrm{d} T} \right)^2 + V_{\mathrm{eff}}(\rho),$$

hvor de to led er hhv. den kinetiske og den potentielle energitæthed. Betydningen af den anden masse $M$ blev 'normaliseret' ud da vi definerede variable. Derfor kommer denne ikke til at indgå i resten af opgaven. I øvrigt udgår betydningen af massen $m$ også (eksplicit i hvert fald), da vi nu snakker om energitæthed, altså energi pr. volumen, og dette ændrer sig ift. hvor stor $m$ er. Dette er udtryk for den simpelhed, som vi tidligere drøftede, at de naturlige enheder giver os. I det klassiske tilfælde har vi at 

$$ V^{CL}_{\mathrm{eff}}(\rho) = - \frac{1}{2 \rho} + \frac{\lambda^2}{2 \rho^2}, $$

mens vi i Generel Relativitetsteori (GR) har 

$$ V^{GR}_{\mathrm{eff}}(\rho) = - \frac{1}{2 \rho} + \frac{\lambda^2}{2 \rho^2} - \frac{\lambda^2}{2 \rho^3}. $$

Igen er ændringen fra CL til GR en lille rettelse ligesom i tyngdekraftsformlen beskrevet i introduktionen. I dette afsnit vil vi fokusere på $V_{\mathrm{eff}}$.
Den potentielle energi er en størrelse som man kan tænke på som en 'indeholdt' energi. Hvor meget energi vil blive udløst hvis vi lader tiden rulle? I vores kontekst vil det sige at jo længere væk fra $M$ som $m$ befinder sig, og jo mere stile den står jo mere potentiel energi, da $M$ kan udøve en kraft og lave den potentielle energi om til kinetisk energi, indtil at $m$ når $R_S$.

### Spørgsmål 2
#### 2 a)
Vi vil her definere en funktion, som finder den potentielle energitæthed af $m$ givet $\rho$ og $\lambda$. Vi vil yderligere give muligheden for at vi differentiere mellem om vi er i den klassiske grænse eller i Generel Rel, hvor GR vil være standarden.

```python
# Funktion til at beregne den potentielle energitæthed i det klassiske tilfælde (GR = False) og GR-tilfældet (GR = True)
def VeffFind(rho, lam, GR = True):
    if GR == False:
        a = 0
    else:
        a = 1
    Veff = -1/(2*rho) + lam**2/(2*rho**2) - a*(lam**2/(2*rho**3))
    return Veff
```

#### 2 b)
Vi vil nu kigge på de lokale ekstrema for den potentielle energitæthed i både det klassiske tilfælde og generel rel.
En simpel måde at finde ekstremaer for en funktion, er at kigge på for hvilke værdier den afledede funktion er lig med $0$. Altså hvis

$$\frac{\mathrm{d} \, V_{\mathrm{eff}}^{CL}}{\mathrm{d} \rho} = \frac{1}{2 \rho^2} - \frac{\lambda^2}{2 \rho^3} = 0,$$

har vi at

$$\rho = 2 \lambda^2$$

Endvidere hvis

$$\frac{\mathrm{d} \, V_{\mathrm{eff}}^{GR}}{\mathrm{d} \rho} = \frac{1}{2 \rho^2} - \frac{\lambda^2}{2 \rho^3} + \frac{3 \lambda^2}{2 \rho^4} = 0,$$

har vi at

$$\rho = \lambda \left(\lambda \pm \sqrt{\lambda^2 - 3} \right)$$

Det kan tydeligt ses at der i det klassiske tilfælde kun er et ekstremum for $\rho \in \mathbb{R}^+$, og at der i GR-tilfældet er to. Yderligere kan vi se, at når $- \sqrt{3} < \lambda < \sqrt{3}$ i GR-tilfældet, så har vi kvadratroden af et negativt tal, og vores afledede funktion er udefineret. Vi kan også se at der ikke eksisterer sådan en grænse i det klassiske tilfælde. Jeg har svært ved at komme med en direkte fortolkning af disse grænser.


Det er også muligt at finde lokale ekstrema numerisk i python. Vi definerer nu en funktion, som givet et interval af $\Delta\rho$ og de tilsvarende potentielle energitætheder $V_{\mathrm{eff}}(\Delta\rho)$ returnerer $\rho_{\mathrm{ekstremum}}$ og $V_{\mathrm{eff}}(\rho_{\mathrm{ekstremum}})$

```python
from numpy import less, greater
from scipy.signal import argrelextrema
def extrema(drho, Vdrho):
    mins = argrelextrema(Vdrho, less)[0]      # Find minimumer.
    if len(mins) > 0:                         # Tjek om et minimum findes.
        rho_min_i = mins[0]                   # Find indexet af rhoet der svarer til lokalt minimum.
        rho_min = drho[rho_min_i]             # Find rhoet der svarer til lokalt minimum.
        V_min = Vdrho[rho_min_i]              # Find værdien af lokalt minimum.
    else:                                     # Ellers angiv ingen fundne minimum.
        rho_min = None
        V_min = None

    maxs = argrelextrema(Vdrho, greater)[0]   # Find maximumer.
    if len(maxs) > 0:                         # Tjek om et maximum findes.
        rho_max_i = maxs[0]                   # Find indexet af rhoet der svarer til lokalt maximum.
        rho_max = drho[rho_max_i]             # Find rhoet der svarer til lokalt maximum.
        V_max = Vdrho[rho_max_i]              # Find værdien af lokalt maximum.
    else:                                     # Ellers angiv ingen fundne maximum.
        rho_max = None
        V_max = None
    return rho_min, V_min, rho_max, V_max
```

#### 2 c)
Vi vil her definere en funktion som givet, $\lambda$, en øvre grænse $\rho_{\mathrm{max}}$ på et interval $\Delta\rho = [1, \rho_{\mathrm{max}}]$ og hvorvidt om det er klassisk eller GR-tilfældet, og returnerer et plot af potentialet i intervallet $\Delta\rho$. Vi lader de specificerede værdier være standard, dvs. $\lambda = 2$, og $\rho_{\mathrm{max}} = 30$.

```python
from numpy import linspace
import matplotlib
# Sætter latex-konfiguration
matplotlib.rcParams['mathtext.fontset'] = 'stix'
matplotlib.rcParams['font.family'] = 'STIXGeneral'
import matplotlib.pyplot as plt

def potPlotter(filename, lam = 2, lim = 30, GR = True):
    x = linspace(1,lim,lim*100)
    
    if GR == False:
        y = VeffFind(x, lam, GR = False)
    else:
        y = VeffFind(x, lam)
    rho_min, V_min, rho_max, V_max = extrema(x,y)
        
    fig, ax = plt.subplots(figsize = (10,6))
    ax.axhline(y=0, c='k', lw=0.5,alpha=0.75)

    if rho_max == None:
        ax.plot(x, y, label = '$\mathcal{U}_{eff}^{CL}$', lw=2, color = 'b')
        ax.plot(rho_min, V_min, 'ko', label = f'rho_min: {"%.3f" % rho_min}, V_min: {"%.3f" % V_min}')
    else:
        ax.plot(x, y, label = '$\mathcal{U}_{eff}^{GR}$', lw=2, color = 'r')
        ax.plot(rho_max, V_max, 'bo', label = f'rho_max: {"%.3f" % rho_max}, V_max: {"%.3f" % V_max}')
        ax.plot(rho_min, V_min, 'ko', label = f'rho_min: {"%.3f" % rho_min}, V_min: {"%.3f" % V_min}')
    ax.set_xlabel('rho',fontsize = 15)
    ax.set_ylabel('Potential', fontsize = 15)
    ax.set_xlim(0,lim)
    ax.set_ylim(-.2,.2)
    ax.legend(fontsize = 15)
    ax.grid()
    ax.set_title(f'{filename}', fontsize  =20)
    
    fig.tight_layout()
    # fig.savefig(f'{filename}.png',dpi=300)
    return fig, ax
```

```python
potPlotter('Klassisk potential', GR=False)
potPlotter('GR potential')
plt.show()
```

Det kan tydeligt ses, at der i det klassiske tilfælde kun er et enkelt ekstremum, nemlig et minimum (omkring $\rho = 8$), mens der i GR-tilfældet er to, et lokalt maximum (omkring $\rho = 2$), og et lokalt minimum (omkring $\rho = 6$). De præcise værdier er angivet i plottene.

Vi kan se at for $\rho > 2$ ligner de to potentialer hinanden til forveksling. De er dog vidt forskellige på den anden side af dette domæne. I det klassiske tilfælde stiger potentialet hurtigt mod uendelig, hvorimod i GR-tilfældet gør den præcis det modsatte. Dette skyldes alt sammen at

$$-\frac{\lambda^2}{2 \rho^3} \to - \infty \qquad \text{når} \qquad \rho \to 0$$
og
$$-\frac{\lambda^2}{2 \rho^3} \to 0 \qquad \quad \,\,\,\, \text{når} \qquad \rho \to \infty,$$

som selvfølgelig er det led som ikke er der i det klassiske tilfælde, hvilekt også forklarer at de to potentialer er så vidt forskellige for $0 < \rho < 2$.



## Bevægelsesligninger og orbitaler
I introduktionen havde vi formlen for tyngdekraften fra $M$ på $m$ i Generel Relativitetsteori

$$ \overrightarrow{F}_{GR} = m \overrightarrow{a} = - \frac{G M m}{r^3} \left(1 + \frac{3 l^2}{c^2 r^2} \right) \, \overrightarrow{r}.$$

Med vores dimensionsløse varibale kan denne nu skrives som

$$\overrightarrow{a} = - \frac{\overrightarrow{\rho}}{2 \rho^3} \left(1 + \frac{3 \lambda^2}{\rho^2}\right).$$

Dette giver os de fire koblede differentialligninger 

$$\, \, \mathrm{(i)} \quad \frac{\mathrm{d} \rho_x}{\mathrm{d} T} = v_x \qquad \qquad \qquad \qquad \, \, \, \,\mathrm{(ii)} \quad \frac{\mathrm{d} \rho_y}{\mathrm{d} T} = v_y\qquad \qquad \qquad$$
$$\mathrm{(iii)} \quad \frac{\mathrm{d} v_x}{\mathrm{d} T} = - \frac{\rho_x}{2 \rho^3} \left(1 + \frac{3 \lambda^2}{\rho^2}\right) \qquad \mathrm{(iv)} \quad \frac{\mathrm{d} v_y}{\mathrm{d} T} = - \frac{\rho_y}{2 \rho^3} \left(1 + \frac{3 \lambda^2}{\rho^2}\right)$$

Hvor det angulære moment pr. masse, er givet ved

$$\lambda = \left\lVert\overrightarrow{v} \times \overrightarrow{\rho}\right\rVert = \left\lvert v_x \rho_y - v_y \rho_x \right\rvert.$$

Med disse kan vi finde $m$'s bevægelse ved at løse et begyndelsesværdiproblem givet startbetingelserne

$$\overrightarrow{\rho}(T=0) = \overrightarrow{\rho_0}$$
$$\overrightarrow{v}(T=0) = \overrightarrow{v_0}$$

Dette gør vi ved at bruge scipy pakken til numerisk at udregne positioner og hastigheder til alle tider, men vi er dog nødt til at begrænse udregningen, da der er nogle forskellige scenarier, hvor det er ufrugtbart at regne videre. Afhængig af hvilke starbetingelser $\overrightarrow{\rho_0}$ og $\overrightarrow{v_0}$ vi har kan vi ende i 3 scenarier:

- **Radiel dyk**

Her ender $m$ inden for $M$'s Schwarzschild-radius, og videre beregning er således irrelevant, da vi ved at $m$ ikke kan undslippe igen. Her definerer vi en funktion som returnerer $m$'s afstand fra $R_S$, så vi kan se om denne grænse nåes i den numeriske løsningsproces. Vi er altså interesserede i grænsen

$$\left\lVert\overrightarrow{\rho}(T)\right\rVert < R_S.$$

- **Sprædningsorbital**

Her ender $m$ med at være så langt væk fra $M$, at tyngdekraften ikke er stor nok til at den kommer tilbage inden for en rimelig tidsgrænse. Her definerer vi en funktion som returnerer $m$'s afstand til 2 gange startpositionen $\overrightarrow{\rho_0}$, så vi kan se om denne grænse nåes i den numeriske løsningsproces. Her er vi altså interesserede i grænsen

$$\left\lVert\overrightarrow{\rho}(T)\right\rVert > 2 \left\lVert\overrightarrow{\rho_0}\right\rVert.$$

- **Bundet orbital**

Her ender $m$ med at være i en stabil orbital i så lang tid, at det ikke er frugtbart at køre udregningen længere. Her sætter vi en tidsgrænse $T_{max}$ på udregningen, så den ikke fortsætter uendeligt. Vi er altså interesserede i tilfældet hvor

$$0 < T < T_{max}.$$


### Spørgsmål 3
#### 3 a)
Først definerer vi en funktion som tager 'tilstanden' af systemet $s = \left(\overrightarrow{\rho}(T), \overrightarrow{v}(T) \right)$ returnere de tidsafledte af disse.

```python
from numpy import sqrt, abs

# Funktion som finder længden af rho givet x,y
def rhoFind(x,y):
    rho = sqrt(x**2 + y**2)
    return rho

# Funktion som finder lambda givet x,y,vx,vy
def lamFind(x,y,vx,vy):
    lam = abs(vx*y - vy*x)
    return lam

# Funktion som tager startbetingelserne (s = [rhox, rhoy, vx, vy]) og beregner de tidsafledte bevægelsesligninger (rhomax, og RS er inkluderet pga 'events' i solve_ivp funktionen)
def diffs(t, s, rhomax, RS):
    x,y,vx,vy=s
    rho = rhoFind(x,y)
    lam = lamFind(x,y,vx,vy)
    K = 1 + (3*lam**2)/rho**2
    
    dx = vx
    dy = vy
    dvx = -x/(2*rho**3)*K
    dvy = -y/(2*rho**3)*K
    return [dx, dy, dvx, dvy]
```

#### 3 b)
Nu definerer vi en funktion, der tager 'tilstanden' af systemet $s$ og returnerer $m$'s afstand fra hhv. $R_S$ og $2 \left\lVert\overrightarrow{\rho_0}\right\rVert$. Vi bemærker at udsagnene

$$\left\lVert \overrightarrow{\rho}(T) \right\rVert < R_S \qquad \text{og} \qquad 2 \left\lVert\overrightarrow{\rho_0}\right\rVert < \left\lVert\overrightarrow{\rho}(T)\right\rVert$$
er ækvivalente med
$$\left\lVert\overrightarrow{\rho}(T)\right\rVert - R_S < 0 \qquad \text{og} \qquad 2 \left\lVert\overrightarrow{\rho_0}\right\rVert - \left\lVert\overrightarrow{\rho}(T)\right\rVert < 0$$

```python
# En funktion der returnerer forskellen mellem rho og RS givet en tid og en systemtilstand.
def dyk(t, s, rhomax, RS = 1):
    x, y, vx, vy=s
    rho = rhoFind(x,y)
    return rho - RS
dyk.terminal = True

# En funktion der returnerer forskellen på rhomax og rho givet en tid og en systemtilstand.
def sprædning(t, s, rhomax, RS = 1):
    x,y,vx,vy=s
    rho = rhoFind(x,y)
    return rhomax - rho
sprædning.terminal = True
```

#### 3 c)
Nu skriver vi en funktion som sammenfatter de sidste par funktioner, til en numerisk løsning af vores begyndelsesværdiproblem ved brug af *scipy.integrate.solve_ivp*-funktionen. Funktionen tager som udgangspunkt systemets begyndelsestilstand $s = [\overrightarrow{\rho_0}, \overrightarrow{v_0}]$. Evt. kan den også tage en største integrationsstid (grænsen for de bundne orbitaler) som er sat til 2000 som standard, en $R_S$-værdi som er sat til 1 og en faktor med hvilken man kan skalere grænsen for sprædningsorbitalerne, som standard er den sat til 2. Funktionen returnerer så alle tiderne $ T = \left\{T_i \right\}$, positionerne $R = \left\{\overrightarrow{\rho}(T_i) | T_i \in T \right\}$, hastighederne $Y = \left\{\overrightarrow{v}(T)| T_i \in T \right\}$ og afstandende $D = \left\{||\overrightarrow{\rho}(T_i)|| \,| \,T_i \in T \right\}$. Yderligere returnerer funktionen også til hvilken tid en af grænserne (radiel dyk eller sprædningsorbital) blev nået. Hvis ikke nogen grænse blev nået returnerer den bare $T_{max}$.

```python
from scipy.integrate import solve_ivp

# Funtkion som returnerer tider, positioner, hastigheder, afstande og integrationstid for orbitaler, givet startværdier og evt en øvre tidsgrænse
def orbitals(s, Tmax = 2000, RS = 1, d=2):
    y0 = x,y,vx,vy = s
    rhomax = d*rhoFind(s[0],s[1])
    t_span = [0,Tmax]
    t_eval = linspace(t_span[0], t_span[1], 100000)
    
    sol = solve_ivp(diffs, t_span = t_span, y0 = y0, t_eval = t_eval, events = [sprædning, dyk], method = 'DOP853', args=(rhomax, RS))
    tim = sol.t
    pos = [sol.y[0], sol.y[1]]
    vel = [sol.y[2], sol.y[3]]
    t_events = sol.t_events
    
    rho = rhoFind(pos[0],pos[1])
    
    if len(t_events[0]) > 0:
        Tmax = t_events[0][0]
    elif len(t_events[1]) > 0:
        Tmax = t_events[1][0]
    else:
        Tmax = Tmax
    return tim, pos, vel, rho, Tmax, sol
```

Nu putter vi vores funktioner i brug, for at beregne orbitalerne i 3 forskellige situationer

##### 3 e) i: Radiel dyk

## Bevarelse af energi og angulært moment 
Et grundprincip i fysik er at i et lukket system (som det vi f.eks. har med at gøre i denne opgave) er energien og angulært moment bevaret. I vores tilfælde vil det sige

$$\frac{\mathrm{d} \mathcal{E}}{\mathrm{d} T} = \frac{\mathrm{d} \lambda}{\mathrm{d} T} = 0.$$

Vi ved at

$$\mathcal{E}= \mathcal{T} + \mathcal{U},$$

hvor $\mathcal{T}$ er den kinetiske del af energitætheden, som er givet ved

$$\mathcal{T} = \frac{1}{2} \left( \frac{\mathrm{d} \rho}{\mathrm{d} T}\right)^2 = \frac{1}{2} \, \frac{\left(\rho_x v_x + \rho_y v_y\right)^2}{\rho_x^2 + \rho_y^2},$$

og $\mathcal{U}$ er den potentielle del af energitætheden, som er givet ved

$$\mathcal{U} = - \frac{1}{2\rho} + \frac{\lambda^2}{2 \rho^2} - \frac{\lambda^2}{2 \rho^3}$$.

Yderligere har vi at

$$\rho = \left\lVert \overrightarrow{\rho} \right\rVert = \sqrt{\rho_x^2 + \rho_y^2}$$

$$ \lambda = \left\lVert \overrightarrow{v} \times \overrightarrow{\rho} \right\rVert = \left\lvert v_X \rho_y - v_y \rho_x\right\rvert$$

Ved at finde disse størrelser og observere hvordan de agerer i løbet af orbitalerne kan vi altså få et dybere indblik i vores beregninger.

### Spørgsmål 4
#### 4 a)
Først vil vi definere en funktion som tager positionerne og hastighederne til alle tidspunkterne i en orbital, og returnere lambda til alle tider. Vi har faktisk lavet en næsten magen til (i **3 a)**), men vi vil gerne have at denne tager positionerne og hastighederne som vektorer i stedet. 

```python
from numpy import abs
# Funktion som finder lambda givet positions- og hastighedsvektor
def lamFindVec(pos,vel):
    x,y = pos
    vx,vy = vel
    lam = abs(vx*y - vy*x)
    return lam
```

#### 4 b)
Som sagt vil vi også gerne observere energitætheden, så nu skriver 2 funktioner der udregner $\rho$ og $\mathcal{T}$ givet vektorer. Igen har vi faktisk lavet to næsten magen til, men vi vil gerne have at de tager positionerne og hastighederne som vektorer i stedet for enkelte koordinater. Hernæst skriver vi en funktion som returnerer den kinetiske- og potentielle energitæthed og deres sum til hvert tidspunkt og den totale energitæthed summeret over hele orbitalen.

```python
# Funktion som finder rho givet positionsvektor
def rhoFindVec(pos):
    x = pos[0]
    y = pos[1]
    rho = sqrt(x**2 + y**2)
    return rho

# Funktion som finder den kinetiske energitæthed givet positions- og hastighedsvektor
def kinFindVec(pos,vel):
    x,y = pos
    vx,vy = vel
    kin = (1/2)*((x*vx + y*vy)**2/(x**2 + y**2))
    return kin

# Vi bruger igen VeffFind og LamFindVec funktionerne defineret tidligere

# Funktion der beregner energitætheden til hvert tidspunkt, og den totale energitæthed, givet positions- og hastighedsvektorer
def eDensFindVec(pos, vel, GR = True):
    rho = rhoFindVec(pos)
    lam = lamFindVec(pos,vel)
    kin = kinFindVec(pos,vel)
    
    if GR == False:
        Veff = VeffFind(rho,lam, GR = False)
    else:
        Veff = VeffFind(rho,lam, GR = True)
        
    E = kin + Veff
    Etot = sum(kin) + sum(Veff)
    
    return kin, Veff, E, Etot
```

#### 4 c)
For at kunne observere de nyligt fundne størrelser, er det nødvendigt at have et plot. Vi skriver derfor en funktion som tager $\mathcal{T}$, $\mathcal{U}$, $\mathcal{E}$ og $\lambda$ og tegner et plot af disse. Vores funktion kan også tage xmax, ymin og ymax, for at vi kan tilpasse plotsne.

```python
from numpy import arange
#Funktion der tegner de forskellige energitæthedstyper og lambda 
def energyPlotter(kin, Veff, E, lam, filename, xmax = 10000, ymin = -0.05, ymax = 0.05):
    t = arange(len(kin))
    fig, ax = plt.subplots(figsize=(14,7))

    ax.plot(t, kin, label = '$\mathcal{T}$', c='r')
    ax.plot(t, Veff, label = '$\mathcal{U}$', c='b')
    ax.plot(t, E, label = '$\mathcal{E}$', c='g')
    ax.legend(loc = 'lower left',bbox_to_anchor=(1.1,0), fontsize=15)
    ax.set_xlabel('T', fontsize = 15)
    ax.set_ylabel('Energitæthed', fontsize = 15)
    ax.set_xlim(0, 50*xmax)
    ax.set_ylim(ymin, ymax)
    ax.set_title(f'{filename}', fontsize = 20)

    ax2 = ax.twinx()
    ax2.plot(t, lam, '--', label = '$\lambda$',c='orange')
    ax2.legend(loc = 'upper left', bbox_to_anchor=(1.1,1), fontsize=15)
    ax2.set_ylabel('Angulært moment', fontsize = 15)
    ax2.set_xlim(0, 50*xmax)
    ax2.set_ylim(min(lam)-0.005,max(lam)+0.005)
    # fig.savefig(f'{filename}.png', dpi=300)
```

Den kinetiske energi starter tæt på $0$, hvilket giver mening, da denne afhænger af farten, som ikke er særlig stor i starten.
