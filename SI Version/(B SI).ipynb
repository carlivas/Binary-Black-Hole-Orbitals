```python
# CM Bound orbital, faraway starting pos
m_1 = 1*c**2/G
m_2 = 1e6*c**2/G

R_S = 2*(m_1+m_2)*G/c**2

x_1 = np.array([0.0, 1000.0])*R_S
x_2 = np.array([0.0,    0.0])*R_S
v_1 = np.array([0.01,   0.0])*c
v_2 = np.array([0.0,    0.0])*c

s1 = x_1, x_2, v_1, v_2, m_1, m_2

pos1, pos2, pos_CM = run_model(s1, t_max = 115000*R_S/c, dt = 100*R_S/c, EIH = False)

# Change to CM frame
pos1 = (pos1 - pos_CM)/R_S
pos2 = (pos2 - pos_CM)/R_S
pos_CM = np.array([0,0])

start_pos = np.array([pos1[0][0], pos1[1][0]])
end_pos = np.array([pos1[0][-1], pos1[1][-1]])

fig, ax = plt.subplots(figsize=(11,11))

xlim, ylim = plotLimsTwoBody(pos1, pos2)
ax.set_xlim(xlim)
ax.set_ylim(ylim)
ax.plot(pos1[0],pos1[1], 'w', label = 'pos1',linewidth=0.75)
ax.plot(pos2[0],pos2[1], 'w', label = 'pos2',linewidth=0.75)
ax.plot(pos_CM[0],pos_CM[1], 'ow', label = 'posCM')
ax.plot(start_pos[0], start_pos[1], 'wx', label = f'Start pos: ({"%.1f" % start_pos[0]}, {"%.1f" % start_pos[1]})')
ax.plot(end_pos[0],end_pos[1], 'w+', label = f'End pos: ({"%.1f" % end_pos[0]}, {"%.1f" % end_pos[1]})')
ax.set_xlabel('$R_S$', fontsize = 15)
ax.set_ylabel('$R_S$', fontsize = 15)
ax.set_facecolor('k')
ax.grid(c='grey', alpha=0.2, ls ='--')
ax.set_aspect('equal')
ax.set_title('CM Bound - Faraway', fontsize = 20)
ax.legend(facecolor='grey',loc='upper right', fontsize = 10)
fig.tight_layout()

```
