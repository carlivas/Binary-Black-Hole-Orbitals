```python
import numpy as np
import matplotlib.pyplot as plt

from matplotlib.animation import FuncAnimation
from IPython.display import HTML
xmin, xmax = -4, 4
r_min, r_max = -4, 4

# Set up the figure

fig = plt.figure(figsize=(8,6))
ax = plt.axes(xlim=(xmin,xmax))

line, = ax.plot([],[], lw=2)

def init():
    line.set_data([], [])
    return line,

# Animation of -4<r<4
def animate(r):
    x = np.linspace(r_min, r_max, 100)
    y = 1+r*x+x**2
    line.set_data(x,y)
    line.set_label('$r={}$'.format(round(r,2)))
    plt.axhline(0, alpha=0.6, c='grey')
    plt.axvline(0, alpha=0.6, c='grey')
    plt.legend(loc='upper left')
    return line,

bifurcation = FuncAnimation(fig, animate, init_func=init, frames=np.linspace(r_min, r_max, 250), interval=6, blit=True)
HTML(bifurcation.to_html5_video())
```
